```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset("penguins")
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
sex = pd.get_dummies(data.sex,drop_first=True)
```

```python
sex.head()
```

```python
data.island.unique()
```

```python
island = pd.get_dummies(data.island)
```

```python
island
```

```python
data = pd.concat([data,island,sex],axis = 1)
```

```python
data.head()
```

```python
data.drop(['island','sex'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.species.unique()
```

```python
#usinf map function to convert categorical values into numeric
Y = data['species']
Y = Y.map({'Adelie':0,'Chinstrap':1,'Gentoo':2})
Y.head()
```

```python
data.drop(['species'],inplace=True,axis = 1)
```

```python
data
```

```python
data = pd.concat([data,Y],axis=1)
```

```python
data
```

```python
X = data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 5,criterion = 'entropy',random_state=0)
```

```python
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report,accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```
